** Reading Text from a File **

```python
file = open('Cambridge_Anlaytics.txt', 'r') 
text = file.read() 
text.encode('utf-8', 'ignore')
```

```python
text
```

** Preprocessing Steps **

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from string import punctuation
```

```python
# Tokenizing into sentences
sents = sent_tokenize(text)
sents
```

```python
# Tokenizing into words
word_sent = word_tokenize(text.lower())
word_sent
```

```python
_stopwords = set(stopwords.words('english') + list(punctuation) + ['—', '``', "'s", "''"])
_stopwords
```

```python
# Removing stopwords
word_sent = [word for word in word_sent if word not in _stopwords]
word_sent
```

```python
from nltk.probability import FreqDist

# Calculating the frequency of occurence for every word in the article
freq = FreqDist(word_sent)
freq
```

```python
from heapq import nlargest
```

```python
# Getting the top 10 words in the article

nlargest(10, freq, key = freq.get)
```

```python
from collections import defaultdict
```

```python
# Calculating the significance score per sentence

ranking = defaultdict(int)

for i, sent in enumerate(sents):
    for w in word_tokenize(sent):
        if w in freq:
            ranking[i] += freq[w]
            
ranking
```

```python
# Getting the index of top significant sentences in the article

sent_idx = nlargest(4, ranking, key = ranking.get)

sent_idx
```

```python
# Getting the sentences from the article that are top significant using sent_idx

[sents[j] for j in sorted(sent_idx)]
```
